### Indian covid cases

```python
india=country.get_group('India')
india
```

### Total number of death in india

```python
sum1=india['deaths'].sum()
```

### Total number of reported Covid cases in India

```python
cases=india['cases'].sum()
```

```python
plt.figure(figsize=(8, 8))
plt.pie([sum1,cases], labels=['Cases', 'Deaths'], autopct='%1.1f%%', colors=['red','skyblue'])
plt.title('Proportion of Total COVID-19 Cases and Deaths in India')
plt.show()
```

## Afghanistan covid cases and deaths

```python
country=covid.groupby(["countriesAndTerritories"])
```

```python
afg=country.get_group('Afghanistan')
```

```python
total_cases = afg['cases'].sum()
total_deaths = afg['deaths'].sum()
plt.figure(figsize=(8, 8))
plt.pie([total_cases, total_deaths], labels=['Cases', 'Deaths'], autopct='%1.1f%%', colors=['linen', 'red'])
plt.title('Proportion of Total COVID-19 Cases and Deaths in Afghanistan')
plt.show()
```
