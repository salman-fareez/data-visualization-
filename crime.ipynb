```python
year_data = crime[crime['Year'] == 2021]
plt.figure(figsize=(12, 6))
sns.barplot(x='Rape',y='State',data=year_data,)
plt.title('Total Rape Cases by State (2021)')
plt.xlabel('Number of Rape Cases')
plt.ylabel('States')
plt.show()
```

```python
andhra= crime[crime['State'] == 'ANDHRA PRADESH']
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Rape', data=andhra, marker='o')
plt.title('Rape Cases Over the Years in Andhra Pradesh')
plt.xlabel('Year')
plt.ylabel('Number of Rape Cases')
plt.show()
```
