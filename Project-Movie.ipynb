# About dataset

Unnamed: 0: Likely an index column or a redundant ID column.
id: The unique identifier for each movie or data entry.
budget: The allocated budget for the movie, likely in millions.
popularity: A measure of how popular the movie is, based on factors like views, mentions, etc.
revenue: The total earnings or revenue the movie generated, possibly in millions.
title: The title or name of the movie.
vote_average: The average rating the movie received, usually out of 10 (scaled to 100 here).
vote_count: The number of people who voted or rated the movie.
director_id: The unique identifier for the movie's director.
year: The year the movie was released or produced.
month: The month the movie was released or produced.
day: The day of the month the movie was released or produced.

## -Q2Christopher Nolan directed movies


```python
cn=imdb.groupby('director_name')
```

```python
nolan=cn.get_group('Christopher Nolan')
nolan
```

```python
plt.figure(figsize=(15,10))
plt.pie(nolan['popularity'], labels=nolan['title'],explode=(0,0,0.05,0,0,0,0,0), autopct='%1.1f%%', startangle=140,colors=['forestgreen','olive','rosybrown','greenyellow','indigo','gold','tomato','maroon'
])
plt.title('popularity of Christopher Nolan  Movies')
plt.show()
```

## -Q7 The top 5 highest budget movies on the list

```python
top=imdb[['id','budget','popularity','revenue','title','vote_average','vote_count','director_id','year','month','day','director_name','gender']].sort_values(by='budget',ascending=False).head(5)
top
```

```python
plt.figure(figsize=(8,8))
plt.pie(top['budget'], labels=top['title'],explode=(0.1,0,0,0,0), autopct='%1.1f%%', startangle=140)
plt.title('Budget Distribution of Top 5 Movies')
plt.show()
```

## -Q8 top 10 highst vote_average get movies

```python
top_movies=imdb[['id','budget','popularity','revenue','title','vote_average','vote_count','director_name','gender']].sort_values(by='vote_average',ascending=False).head(10)
top_movies
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_movies['title'], top_movies['vote_average'], color='lightsalmon')
plt.xlabel('vote_average)')
plt.title('Top 10 Movies by avarge vote')
plt.show()
```

## -Q10 top 10 popular movies

```python
top10=imdb[['budget','popularity','revenue','title','director_name','gender']].sort_values(by='popularity',ascending=False).head(10)
top10
```

```python
plt.figure(figsize=(10, 6))
plt.plot(top10['title'],top10['revenue'], marker='o', linestyle='-')
plt.title('Profit of Top 10 Movies by Popularity')
plt.xlabel('title')
plt.ylabel('revenue')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()

```
